```python
import numpy as np
import pandas as pd 
```

```python
# import website.csv into a dataframe called whiteWine. 
path = "C:\\Users\\ptkar\\Desktop\\Python Data Science\\website.csv"
df = pd.read_csv(path, sep=",")

df.head()
```

```python
# Turn Charset and Server features into dummy variables

dummy_features_gen = pd.get_dummies(df['CHARSET'], drop_first=False)

df = pd.concat([df,dummy_features_gen], axis =1 )
df = df.drop('CHARSET', axis = 1)

dummy_features_gen = pd.get_dummies(df['SERVER'], drop_first=False)
df = pd.concat([df,dummy_features_gen], axis =1 )
df = df.drop('SERVER', axis = 1)



# removing url 
df = df.drop('URL', axis = 1)
```

```python
# turning 'None' values into 01/01/1990
type(df['WHOIS_REGDATE'][0])

df['WHOIS_REGDATE'] = df['WHOIS_REGDATE'].apply(lambda x: '01/01/1900 0:00' if x == 'None' else x )
df['WHOIS_UPDATED_DATE'] = df['WHOIS_UPDATED_DATE'].apply(lambda x: '01/01/1900 0:00' if x == 'None' else x )

# syntex for splitting stings on a DF came from https://stackoverflow.com/questions/14745022/how-to-split-a-column-into-two-columns
# purpose of this is to get rid of the time field as I am looking to see many days we currently am from this date.
# for furthur analysis, it might be worth binning the times into morning/afternoon/night.

df['WHOIS_REGDATE'] = df['WHOIS_REGDATE'].str.split(' ',1,expand=True)[0]
df['WHOIS_UPDATED_DATE'] = df['WHOIS_UPDATED_DATE'].str.split(' ',1,expand=True)[0]

```

```python
# changing the Whois_country and whois_StateProd into binary variables
df['WHOIS_COUNTRY'] = df['WHOIS_COUNTRY'].apply(lambda x: 0 if x == 'None' else 1 )
df['WHOIS_STATEPRO'] = df['WHOIS_STATEPRO'].apply(lambda x: 0 if x == 'None' else 1 )
```

```python
# Turning NaN values to 0
df['CONTENT_LENGTH'] = np.nan_to_num(df['CONTENT_LENGTH'])

df['DNS_QUERY_TIMES'] = np.nan_to_num(df['DNS_QUERY_TIMES'])

```

```python
# Droping whois_regdate and whois_updated_Date due to formatting issues. Would like to re-visit though if time permits.
df = df.drop('WHOIS_REGDATE', axis = 1)
df = df.drop('WHOIS_UPDATED_DATE', axis = 1)
```
